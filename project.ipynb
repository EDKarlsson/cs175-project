```python
import pickle
import os
from src import synthesizer
import keras
import numpy as np

CURRENT_DIR = os.getcwd()
DATA_DIR = CURRENT_DIR.split('cs175-project')[0] + "cs175-project/data"
```

### Show the classification of the real and fake articles

```python
learner = pickle.load(open(DATA_DIR + "/logistic_regression.model", "rb"))
x_test = pickle.load(open(DATA_DIR + "/x_test_sample.txt", "rb")) # Generated Articles
y_test = pickle.load(open(DATA_DIR + "/y_test_sample.txt", "rb"))
unshuffled_set = pickle.load(open(DATA_DIR + "/unshuffled_test_set.project", "rb"))
```

### Bag of word vector for the corresponding fake article from LSTM
### Note: Class 1 is fake, 0 is real

```python
pred = learner.predict([unshuffled_set[corresponding_article]])
print("Fake Article, Model Predicts Class:", pred)
```

```python
pred = learner.predict([unshuffled_set[101+corresponding_article]])
print("Real Article, Model Predicts Class:", pred)
```

### Prediction on sample from the real test set which is a small subset.

```python
# Shuffled Data
score = learner.score(x_test, y_test)
print("Score", score)
```
